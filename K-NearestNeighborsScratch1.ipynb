```python
def euclidean_dist(x1,x2):
    euc_dist = 0
    for ind in range(len(x1)):
        sq_dist = pow((x1[ind]-x2[ind]),2)
        euc_dist = euc_dist + sq_dist
    return (math.sqrt(euc_dist))
```

```python
def nearest_neighbors(x_train, x_test, k):
    distances = []
    neighbors = []
    for train_row in x_train:
        distances.append((train_row, euclidean_dist(train_row, x_test)))
    distances.sort(key = lambda tup: tup[1])
    top_k = distances[0:k]
    return top_k
```

```python
def knn_predict(x_train, x_test, k):
    neighbors = nearest_neighbors(x_train, x_test, k)
    top_neighbors = [x[-1] for x in neighbors]
    pred_outcome = max(set(top_neighbors), key=top_neighbors.count)
    return pred_outcome
```
