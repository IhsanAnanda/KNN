```python
import pandas as pd
import math
import numpy as np
import operator
from sklearn.metrics.pairwise import manhattan_distances
```

```python
#Load dataset
irisDataset = pd.read_csv("iris.csv", sep=',', skiprows=0)
irisDataset
```

```python
#Memisahkan atribut dengan target class
X = irisDataset.iloc[:, 1:6].values
y = irisDataset['Species'].values
print("Atribut Dataset :")
print(X)
```

```python
# Euclidean Distance
def euclideanDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
        distance += pow((instance1[x] - instance2[x]), 2)
    return math.sqrt(distance)
```

```python
# Testing Eucdistances 
data1 = [2, 2, 2]
data2 = [4, 4, 4]
distance = euclideanDistance(data1, data2, 3)
print ("Distance: ",distance)
```

```python
# Manhattan Distances
# data1 = [[2, 3, 2], [4, 4, 4]]
data1 = [[2, 2, 2]] # data test
data1 = np.array(data1)
print(data1.shape)
data2 = [[4, 4, 4], [3, 3, 3]] # data training
data2 = np.array(data2)
print(data2.shape)
distance1 = manhattan_distances(data1, data2)
print("Data Asli : ",distance1)
distances = distance1.reshape((2,1))
print("Data Hasil Reshape :", distances)
```

```python
# KNN
def getKNeighbors(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    distances = np.array(distances, dtype=object) # Mengubah ke numpy array agar mudah diakses
    print(distances.shape) 
    for data in distances:
        print(data[1]) # Hasil Euclidean Distance [1]
    return neighbors
```

```python
# KNN Testing
# trainSet = [[2, 2, 2, 'a'], [4, 4, 4, 'b']]
# trainSet = np.array(trainSet, dtype=object)
trainSet = X
testInstance = [5.6, 2.7, 4.2, 1.3]
testInstance = np.array(testInstance, dtype=object)
k = 5
print(testInstance[0])
neighbors = getKNeighbors(trainSet, testInstance, k)
print("\n",neighbors[0]) # Ukuran array neighbors bergantung pada nilai k
```

```python
# Classification Result
def getResponse(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
```

```python
# neighbors = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]
print(getResponse(neighbors))
```

```python
def getAccuracy(testSet, predictions):
    correct = 0
    for x in range(len(testSet)):
        if testSet[x][-1] is predictions[x]:
            correct += 1
    return (correct/float(len(testSet))) * 100.0
```

```python
testSet = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]
predictions = ['a', 'a', 'b']
accuracy = getAccuracy(testSet, predictions)
print(accuracy)
```
